# Data Cleaning


```python
# Import thư viện
import pandas as pd
import numpy as np

# Mở bộ dữ liệu
data_link = 'https://docs.google.com/spreadsheets/d/1vc6tQnfyYVXJnLp29p--mFqdmXNp-W5r/edit?usp=sharing&ouid=104159724422636186932&rtpof=true&sd=true'

data_path = 'https://drive.google.com/uc?export=download&id=' + data_link.split("/")[-2]

df = pd.read_excel(data_path)

# In bộ dữ liêu
df
```

```python
# Kiểm tra một số thông tin cơ bản của bộ dữ liệu
df.info()
```

```python
# Mô tả chung về dữ liệu
df.describe()
```

```python
# Loại bỏ các cột không có giá trị phân tích
df = df.drop(columns = 'Unnamed: 0')

# Loại bỏ các dòng lặp, giữ lại dòng xuất hiện cuối cùng đối với các dòng  bị lặp
duplicate_check = df.drop_duplicates(keep='last')

# Thay đổi kiểu dữ liệu của cột Retailer ID
df['Retailer ID'] = df['Retailer ID'].astype(object)
```

```python
# Kiểm tra giá trị cột Total Sales
df['total_sales'] = df['Price per Unit'] * df['Units Sold']
df_error_total_sales = df.loc[df['total_sales'] != df['Total Sales']]
df_error_total_sales
```

```python
# Kết quả kiểm tra trả về 3886 dòng có kết quả tính Total Sales không chính xác
# Tương tự, kết quả lợi nhuận của 3886 dòng trên không chính xác
# Sử dụng cột total_sales để tiếp tục tính toán và  phân tích

# Loại bỏ cột Total Sales
df = df.drop(['Total Sales'], axis=1)

# Đổi tên cột total_sales thành Total Sales
df.rename(columns = {'total_sales':'Total Sales'}, inplace = True)

df
```

```python
# Tính lại giá trị cột Operating Profit dựa trên giá trị cột Operating Margin

# Loại bỏ cột Operating Profit gốc
df = df.drop(['Operating Profit'], axis=1)

# Bổ sung cột Operating Profit mới
df['Operating Profit'] = df['Total Sales'] * df['Operating Margin']

df
```

```python
# Bổ sung thêm cột dùng cho phân tích chi tiết
# Bổ sung cột Category bằng hàm map()
category_mapping = {
    "Men's Street Footwear": "Street Footwear",
    "Men's Athletic Footwear": "Athletic Footwear",
    "Men's Apparel": "Apparel",
    "Women's Street Footwear": "Street Footwear",
    "Women's Athletic Footwear": "Athletic Footwear",
    "Women's Apparel": "Apparel"
}

df['Category'] = df['Product'].map(category_mapping)

df
```

```python
# Bổ sung cột Gender bằng hàm map()
gender_mapping = {
    "Men's Street Footwear": "Male",
    "Men's Athletic Footwear": "Male",
    "Men's Apparel": "Male",
    "Women's Street Footwear": "Female",
    "Women's Athletic Footwear": "Female",
    "Women's Apparel": "Female"
}

df['Gender'] = df['Product'].map(gender_mapping)

df
```

```python
# Kiểm tra lại thông tin bộ dữ liệu
df.info()
```

```python
# Kiểm tra lại mô tả bộ dữ liệu
df.describe()
```
